```python
import json
import tiktoken

enc = tiktoken.encoding_for_model("gpt-3.5-turbo")

def count_tokens_from_jsonl(file_path):
    total_tokens = 0

    with open(file_path, "r", encoding="utf-8") as f:
        for line in f:
            if not line.strip():
                continue
            data = json.loads(line)

            if "messages" in data:
                for msg in data["messages"]:
                    if "content" in msg:
                        text = msg["content"]
                        tokens = len(enc.encode(text))
                        total_tokens += tokens

    return total_tokens

file_path = "/content/gdrive/MyDrive/DILAB/YB/GPT_FineTuning/GPT-3.5-turbo/finetune_data.jsonl"
tokens = count_tokens_from_jsonl(file_path)
print(f"총 토큰 수: {tokens}")
print(f"예상 학습 비용(3 epoch 기준): ${tokens * 3 / 1000 * 0.008:.2f}")
```
